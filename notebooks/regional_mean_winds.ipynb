This notebook demonstrates calculation to estimate regional mean u and v winds for the Bering Shelf, East Siberian Sea, and Siberian Shelf regions.  The code uses xarray.

```python
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

import matplotlib.patches as mpatches
```

The first step is to define the regions we need for the calculations.  These are defined using list objects [lat0, lat1, lon0, lon1]

```python
ess_region = [70., 80., 140., 180.]   # East Siberian Shelf
bss_region = [65., 67., -168., -166.] # Bering Sea Shelf
sis_region = [70., 80., 100., 180.]   # Siberian Shelf
```

```python
fig = plt.figure(figsize=(7, 7))
    
# Set up map projection and extent
map_proj = ccrs.NorthPolarStereo(central_longitude=180.)
extent = [150.,210.,40.,90.]
    
ax = plt.subplot( projection=map_proj )
ax.set_extent( extent, ccrs.PlateCarree())

#ax.add_feature( cfeature.LAND )
ax.coastlines()
ax.gridlines()
x0 = ess_region[2]
y0 = ess_region[0]
w0 = ess_region[3]-ess_region[2]
h0 = ess_region[1]-ess_region[0]
ax.add_patch(mpatches.Rectangle(xy=[x0, y0], width=w0, height=h0,
                                facecolor='red',
                                alpha=0.2,
                                transform=ccrs.Geodetic()) )

x1 = bss_region[2]
y1 = bss_region[0]
w1 = bss_region[3]-bss_region[2]
h1 = bss_region[1]-bss_region[0]
ax.add_patch(mpatches.Rectangle(xy=[x1, y1], width=w1, height=h1,
                                facecolor='blue',
                                alpha=0.2,
                                transform=ccrs.Geodetic()) )
  
plt.show()
```
