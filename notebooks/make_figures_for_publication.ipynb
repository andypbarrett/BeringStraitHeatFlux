# Make figures for publication

The notebook generates the figures for the Bering Strait Oceanic volume and heat transport paper submitted to JGR-Oceans.  The purpose of the notebook is to document the input files and plotting routines used to generate figures.  This should save a lot of hunting around and trying to remember script names.  It will also allow figure to be modified.

Source code for figure generation will be collected in a module in `../source`.  This will save lots of code appearing in the notebook.

*At the moment, the notebook is a work in progress but it could be turned into an interactive paper*

````python
import os
import xarray as xr

import sys
sys.path.append('../source')

import figures

from constants import REANALYSIS_DIRPATH, MOORING_DIRPATH

FIGURES_PATH = '../figures'
````

## Figure 04

Composite sea level pressure anomaly fields for the 15 largest and the 15 smallest PHtrans270Bering Strait for winter (December through February, DJF), spring (March through May, MAM), summer 271(June through August, JJA) and autumn (September through November, SON) and the composite 272differences (high minus low).  Regions with significant pressure anomalies at the 95% level are shown in 273color.  Non-significant regions are in white. Anomalies are computed with respect to averages for each 274season over the period 1990-2016.

Composites are created with ```make_composites_for_season```.  A shell script *name of shell script* is in the same directory.

````python
slp_composite_file = 'slp_by_ph_composites_for_season.nc4'
u10m_composite_file = 'u10m_by_ph_composites_for_season.nc4'
v10m_composite_file = 'v10m_by_ph_composites_for_season.nc4'

slp = xr.open_dataset(os.path.join(REANALYSIS_DIRPATH, 'SLP', slp_composite_file))
u10m = xr.open_dataset(os.path.join(REANALYSIS_DIRPATH, 'U10M', u10m_composite_file))
v10m = xr.open_dataset(os.path.join(REANALYSIS_DIRPATH, 'V10M', v10m_composite_file))

fig04 = figures.plot_composite_figure(slp, u10m, v10m)

fig04.savefig(os.path.join(FIGURES_PATH, 'bering-strait-heat_figure04_composites-by-ph.png'))
````
